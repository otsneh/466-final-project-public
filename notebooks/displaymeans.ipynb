```python
import numpy as np
import matplotlib.pyplot as plt
def displaymeans(means):

    z = []
    z = means

    fig, ax = plt.subplots()
    im = ax.imshow(z)
    fig.colorbar(im)
    for i in range(len(z)):
        for j in range(len(z[1])):
            text = ax.text(j, i, round(z[i][j],1),
                           ha="center", va="center", color="w")
    ax.axis('tight')
    fig.tight_layout()
    plt.show(ax)
```

```python
z = [[1,1],[1,1]]
displaymeans(z)
```
