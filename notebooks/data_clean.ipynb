```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./alsfrs.csv')
```

```python
df.head()
```

```python
# merge Q5a and Q5b

df['Q5_Cutting'] = df['Q5a_Cutting_without_Gastrostomy'].fillna(df['Q5b_Cutting_with_Gastrostomy'])
```

```python
df.head()
```

```python
# Fill Q10_Respitory

df['Q10_Respiratory'] = df['Q10_Respiratory'].fillna((df['R_1_Dyspnea'] + df['R_2_Orthopnea'] + df['R_3_Respiratory_Insufficiency']) // 3)
```

```python
df.head()
```

```python
# Fill ALRFS_Total (old gets new)
df['ALSFRS_Total'] = df['ALSFRS_Total'].fillna(df['ALSFRS_R_Total'])
```

```python
df.iloc[:30].dropna()
```

```python
# drop Mode_of_administration and responded by
df = df.drop(columns=['Mode_of_Administration', 'ALSFRS_Responded_By', 
                 'R_1_Dyspnea', 'R_2_Orthopnea', 'R_3_Respiratory_Insufficiency', 
                 'ALSFRS_R_Total', 'Q5b_Cutting_with_Gastrostomy', 'Q5a_Cutting_without_Gastrostomy'])
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.insert(5, 'Q5_Cutting1', df['Q5_Cutting'], allow_duplicates=True)
```

```python
df = df.drop(df.columns[-1], axis=1)
```

```python
df = df.rename(columns={'Q5_Cutting1': 'Q5_Cutting'})
```

```python
def calc_diff(x):
    x['delta_t'] = x['ALSFRS_Delta'].diff()
    return x

# calculate time deltas
df = df.groupby('subject_id').apply(calc_diff)
```

```python
df.to_csv('alsfrs_cleaned.csv')
```
